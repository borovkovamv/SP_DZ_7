## Задание 1

Вернемся к [набору данных о видеоиграх](https://github.com/obulygin/pyda_homeworks/blob/master/stat_case_study/video_games_sales.csv).

Ответьте на следующие вопросы:

1) Как критики относятся к спортивным играм?  
2) Критикам нравятся больше игры на PC или на PS4?  
3) Критикам больше нравятся стрелялки или стратегии?  

Для каждого вопроса:
- сформулируйте нулевую и альтернативную гипотезы;
- выберите пороговый уровень статистической значимости;
- опишите полученные результаты статистического теста.

````python
import pandas as pd
from scipy import stats as st
````

````python
df = pd.read_csv('video_games_sales.csv')
df.info()
df.head()
````

````python
# 1) Как критики относятся к спортивным играм?

df_sport = df[ df['Genre'] == 'Sports']
df_sport = df_sport[['Genre', 'Critic_Score']].dropna()
df_sport
````

````python
# средняя оценка по всем жанрам:
df.Critic_Score.mean()
````

````python
# гипотезы:
# H0: сердняя оценка спортивным играм >= 7, критикам нравятся спортивные игры
# H1: сердняя оценка спортивным играм < 7, критикам не нравятся спортивные игры
# уровень значимости: 0.05


alpha = 0.05
result = st.ttest_1samp(df_sport['Critic_Score'], 7)
print(result)

if (result.statistic < 0) & (result.pvalue / 2 < alpha):
    print('Отвергаем нулевую гипотезу, средняя оценка критиков меньше 7')
else:
    print('Не отвергаем нулевую гипотезу, средняя оценка критиков больше 7')
````

### Вывод: с большой достоверностью мы можем говорить, что спортивные игры критикам нравятся, их средняя оценка выше 7

````python
# 2) Критикам нравятся больше игры на PC или на PS4?

df_platform = df[ (df['Platform'] == 'PC') | (df['Platform'] == 'PS4')]
df_platform = df_platform[['Platform', 'Critic_Score']]
df_platform.head()
````

````python
# гипотезы:
# H0: средняя оценка для игр на PC и PS4 одинаковая
# H1: средняя оценка для игр на PC и PS4 разная
# уровень значимости: 0.05


alpha = 0.05
result = st.ttest_ind(df[ df['Platform'] == 'PC' ]['Critic_Score'].dropna(), df[ df['Platform'] == 'PS4' ]['Critic_Score'].dropna(), equal_var=False)
print(result)

if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу, средняя оценка разная')
else:
    print('Не отвергаем нулевую гипотезу, средняя оценка схожая')
````

### Вывод: с большой долей достоверности можем утверждать, что средние оценки разные и так как statistic < 0, то средняя оценка критиков по играм на PC меньшем, чем на PS4, следовательно критикам игры на PS4 нравятся больше.

````python
# 3) Критикам больше нравятся стрелялки или стратегии?

df_genre = df[ (df['Genre'] == 'Shooter') | (df['Genre'] == 'Strategy')]
df_genre = df_genre[['Genre', 'Critic_Score']]
df_genre
````

````python
# гипотезы:
# H0: средняя оценка для стрелялок и стратегий одинаковая
# H1: средняя оценка для стрелялок и стратегий разная
# уровень значимости: 0.05


alpha = 0.05
result = st.ttest_ind(df[ df['Genre'] == 'Shooter' ]['Critic_Score'].dropna(), df[ df['Genre'] == 'Strategy' ]['Critic_Score'].dropna(), equal_var=False)
print(result)

if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу, средняя оценка разная')
else:
    print('Не отвергаем нулевую гипотезу, средняя оценка схожая')
````

### Вывод: с вероятностью 95% стратегии нравятся критикам так же как и стрелялки

````python
import re
import nltk
````

````python
spam = pd.read_csv('spam.csv')
spam.head()
````

````python
# 1) Привидите весь текст к нижнему регистру
spam['Message'] = [spam.lower() for spam in spam.Message]
# 2) Удалите мусорные символы
spam['Message'] = [re.sub('[\W_]+',' ', text) for text in spam['Message']]
spam['Message'] = [text.split() for text in spam['Message']]
````

````python
nltk.download('stopwords')
nltk.download('wordnet')

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
````

````python
# 3) Удалите стоп-слова
stopwords_set = set(stopwords.words('english'))
spam['Message'] = spam['Message'].apply(lambda x: [word for word in x if word not in stopwords_set])
# 4) Привидите все слова к нормальной форме
spam['Message'] = spam['Message'].apply(lambda x: [WordNetLemmatizer().lemmatize(word) for word in x])
````

````python
spam['Message'] = spam['Message'].apply(lambda x: ' '.join(x))
spam.head()
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer
````

````python
# 5) Преобразуйте все сообщения в вектора TF-IDF
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(spam['Message'])
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
tfidf_matrix
````

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
````

````python
# 6) Разделите данные на тестовые и тренировочные в соотношении 30/70, укажите random_state=42. Используйте train_test_split
x_train, x_test, y_train, y_test = train_test_split(tfidf_matrix, spam['Category'], test_size = 0.3, random_state = 42)
````

````python
# 7) Постройте модель логистической регрессии, укажите random_state=42, оцените ее точность на тестовых данных
lr = LogisticRegression(random_state=42)
lr.fit(x_train, y_train)
````

````python
print(f'Точность прогнозирования модели: {round(accuracy_score(y_test, lr.predict(x_test)) * 100)}%')
````

````python
# 8) Опишите результаты при помощи confusion_matrix
confusion_matrix(y_test, lr.predict(x_test))
````

Вывод: ошибка прогнозирования на тестовых данных 69 значений, что равно 4% от 1672, доказывая правильное поведение модели при точности прогнозирования 96% 

````python
# 9) Постройте датафрейм, который будет содержать все исходные тексты сообщений, классифицированные неправильно (с указанием фактического и предсказанного)
spam['Predict category'] = lr.predict(tfidf_matrix)
spam['Message'] = pd.read_csv('spam.csv')['Message']
spam_wrong = spam[ spam['Predict category'] != spam['Category']]
spam_wrong
````
